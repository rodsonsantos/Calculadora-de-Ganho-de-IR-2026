```python
# Caso o projeto seja aprovado, veja quanto você deixará de pagar se seu salário for até R$ 7.350

# Entrada do salário bruto mensal
salario_bruto = float(input('Insira o seu salário bruto mensal para descobrir o impacto da nova regra:'))

# Parâmetros
DESCONTO_SIMPLIFICADO = 607.20
BASE_ISENCAO = 2428.80 # Base de cálculo máxima para isenção (2.428,80)

print(f'\nSalário bruto: R$ {salario_bruto:.2f}')
ganho_mensal = 0.0

# 1. Sem ganho: salário acima de R$ 7.350,00
if salario_bruto >= 7350.00:
    print('Sem ganho mensal com a nova regra para este salário.')

# 2. Faixa especial: R$ 5.000,01 a R$ 7.350,00
elif 5000.01 <= salario_bruto < 7350.00:
    ganho_mensal = 978.62 - (0.133145 * salario_bruto)
    print(f'Ganho mensal: R$ {ganho_mensal:.2f}')
    print(f'Ganho anual (considerando 13º salário): R$ {ganho_mensal * 13:.2f}')

# 3. Faixas normais: salário bruto até R$ 5.000,00 (AQUI JÁ ESTÁ IMPLÍCITO QUE salario_bruto <= 5000.00)
else: 
    base_calculo = salario_bruto - DESCONTO_SIMPLIFICADO
    alíquota = 0
    parcela_a_deduzir = 0

    # Determinação da alíquota e parcela a deduzir (sem usar o limite superior de 5000.00, pois já está no 'else' principal)
    if base_calculo > BASE_ISENCAO: # Se a base for maior que o valor de isenção, calcula imposto
        if base_calculo <= 2826.65:
            alíquota = 0.075
            parcela_a_deduzir = 182.16
        elif base_calculo <= 3751.05:
            alíquota = 0.15
            parcela_a_deduzir = 394.16
        elif base_calculo <= 5000.00: # Base de cálculo máxima para essa simulação
            alíquota = 0.225
            parcela_a_deduzir = 675.49
            
    # Cálculo final do ganho (se for positivo)
    ganho_ir = (base_calculo * alíquota) - parcela_a_deduzir

    if ganho_ir <= 0:
        print('Sem ganho mensal com a nova regra para este salário.')
    else:
        ganho_mensal = ganho_ir
        print(f'Ganho mensal: R$ {ganho_mensal:.2f}')
        print(f'Ganho anual (considerando 13º salário): R$ {ganho_mensal * 13:.2f}')
```
